```python
## input text article
article_text="Just what is agility in the context of software engineering work? Ivar Jacobs"
```

```python
import re
import nltk
from nltk.tokenize import word_tokenize
```

```python
article_text = article_text.lower()
article_text
```

```python
# remove spaces, punctuations and numbers
clean_text = re.sub('[^a-zA-Z]', ' ', article_text)
clean_text = re.sub('\s+', ' ', clean_text)
clean_text
```

```python
import nltk
nltk.download('punkt')
```

```python
# split into sentence list
sentence_list = nltk.sent_tokenize(article_text)
sentence_list
```

```python
tokens=word_tokenize(article_text)
```

```python
tokens
```

```python
import nltk
nltk.download("stopwords")
```

```python
stopwords = nltk.corpus.stopwords.words('english')
word_frequencies = {}
for word in tokens:
 if word not in stopwords:
  if word not in word_frequencies:
    word_frequencies[word] = 1
```

```python
word_frequencies
```

```python
import nltk
from nltk.corpus import brown
from nltk import FreqDist
news_text = brown.words(categories='news')
fdist = FreqDist([w.lower() for w in news_text])
```

```python
word_freq=FreqDist(word_frequencies)
```

```python
word_freq.plot(30,title='freqplot')
```

```python
from collections import Counter
```

```python
dictionary=Counter(word_freq)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
cloud=WordCloud(max_font_size=80,colormap='hsv').generate_from_frequencies(dictionary)
```

```python
plt.figure(figsize=(16,13))
plt.imshow(cloud,interpolation='bilinear')
plt.axis('off')
plt.show() 
```
